```python
csv_path2 = "Gemeinsamer_Bundesausschluss/processed/Qualitaetsbericht_Aortenaneurysmen_rupt_23.csv"

# Beispiel: CSV einlesen
qb = pd.read_csv(csv_path2, dtype={'PLZ': str, 'ARS': str, 'AGS': str})

# Funktion zum Parsen der Koordinaten aus dem String
def parse_coords(s):
    # s sieht so aus: "c(3527073.62751839, 6075050.21766101)"
    # Extrahiere die Zahlen per Regex
    numbers = re.findall(r"[-+]?\d*\.\d+|\d+", s)
    if len(numbers) == 2:
        x, y = map(float, numbers)
        return Point(x, y)
    else:
        return None

# Neue Spalte geometry mit Point-Objekten
qb["geometry"] = qb["geometry"].apply(parse_coords)

# GeoDataFrame mit passendem CRS erstellen (hier ein Beispiel: EPSG:31467)
gdf_points = gpd.GeoDataFrame(qb, geometry="geometry", crs="EPSG:31467")
print(gdf_points)
gdf_points.plot()
```
