```python
# 1. CSV-Datei laden
csv_path = "Gemeinsamer_Bundesausschluss/processed/QB_J_2023_KRS.csv"

qb_KRS = pd.read_csv(csv_path, dtype={'ARS': str})
qb_KRS = qb_KRS.drop('geometry', axis=1)

qb_KRS
```

```python
csv_path2 = "Gemeinsamer_Bundesausschluss/processed/QB_J_2023.csv"

# Beispiel: CSV einlesen
qb = pd.read_csv(csv_path2, dtype={'PLZ': str, 'ARS': str, 'AGS': str})

# Funktion zum Parsen der Koordinaten aus dem String
def parse_coords(s):
    # s sieht so aus: "c(3527073.62751839, 6075050.21766101)"
    # Extrahiere die Zahlen per Regex
    numbers = re.findall(r"[-+]?\d*\.\d+|\d+", s)
    if len(numbers) == 2:
        x, y = map(float, numbers)
        return Point(x, y)
    else:
        return None

# Neue Spalte geometry mit Point-Objekten
qb["geometry"] = qb["geometry"].apply(parse_coords)

# GeoDataFrame mit passendem CRS erstellen (hier ein Beispiel: EPSG:31467)
qb = gpd.GeoDataFrame(qb, geometry="geometry", crs="EPSG:31467")

print(qb.head())
qb.plot()

```

```python
# 1. Shapefile laden
shapefile_path = "C:/Users/agnes/Documents/EAGLE/Innovation_Lab/Daten/Verwaltungsgebiete/vg250_12-31.gk3.shape.ebenen/vg250_ebenen_1231/VG250_KRS.shp"
gdf_KRS  = gpd.read_file(shapefile_path, dtype={'AGS': str, 'ARS': str, 'SDV_ARS': str})
gdf_KRS
# print(gdf_shp.crs)
```

```python
# Dissolve: alle Geometrien mit gleichem SDV_ARS zu einem Polygon zusammenführen
gdf_SDV_ARS_merge = gdf_KRS.dissolve(by='SDV_ARS', as_index=False)

# Ergebnis prüfen
gdf_SDV_ARS_merge
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 10))

# Polygone in Grau mit schwarzem Rand
gdf_KRS.plot(ax=ax, facecolor='lightgray', edgecolor='black')

# Punkte in Rot
qb.plot(ax=ax, color='red', markersize=15)

# Optional: Achsen ausblenden
ax.set_axis_off()

# Optional: Titel setzen
plt.title("Punkte in Polygonen", fontsize=15)

plt.show()
```

```python
# 3. Anzahl der Vorkommen je ARS in der CSV-Datei zählen
ars_counts = qb['ARS'].value_counts().reset_index()
ars_counts.columns = ['ARS', 'Anzahl']
ars_counts
```

```python
#  Join der Daten
gdf_KRS_add = gdf_KRS.merge(qb_KRS, on='ARS', how='left')
gdf_KRS_add
```

```python
# 5. Join der Daten
gdf_FA = gdf_KRS.merge(ars_counts, on='ARS', how='left')
gdf_FA = gdf_FA.rename(columns={'Anzahl': 'Gefaesschirurgische_Fachabteilungen'})
gdf_FA['Gefaesschirurgische_Fachabteilungen'] = gdf_FA['Gefaesschirurgische_Fachabteilungen'].fillna(0)
gdf_FA['Gefaesschirurgische_Fachabteilungen'] = gdf_FA['Gefaesschirurgische_Fachabteilungen'].astype(int)

gdf_FA
```

```python

# Werte sortieren
values = sorted(gdf_FA['Gefaesschirurgische_Fachabteilungen'].unique())

# Colormap erstellen
cmap_base = plt.get_cmap('OrRd', len([v for v in values if v > 0]))
colors = ['lightgrey'] + [cmap_base(i) for i in range(1, len(values))]

# Diskrete Colormap + Norm

cmap = mcolors.ListedColormap(colors)
norm = mcolors.BoundaryNorm(boundaries=np.arange(len(values)+1), ncolors=len(values))

# Plot erstellen
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
gdf_FA.plot(
    column='Gefaesschirurgische_Fachabteilungen',
    cmap=cmap,
    norm=norm,
    linewidth=0.5,
    edgecolor='grey',
    ax=ax
)

# # Punkte plotten
qb.plot(
    ax=ax,
    color='black',
    markersize=15,
    alpha=0.8,
    edgecolor ="white",
    linewidth=0.5
)

# Manuelle Legende erstellen
legend_patches = [
    mpatches.Patch(color=colors[i], label=f"{values[i]}")
    for i in range(len(values))
]

ax.legend(
    handles=legend_patches,
    title='Anzahl Fachabteilungen',
    loc='upper left',
    bbox_to_anchor=(1.02, 0.5),
    fontsize='small',
    title_fontsize='medium',
    frameon=True
)

# Titel & Achsen
ax.set_title("Anzahl gefäßchirurgischer Fachabteilungen je ARS", fontsize=15)
plt.axis('off')
# plt.savefig("Gemeinsamer_Bundesausschluss/processed/Abbildungen/Anzahl_Fachabteilungen_KRS.png", dpi=300, bbox_inches='tight')

plt.show()
# Karte als PNG-Datei speichern

```

```python
# 5. Join der Daten
gdf_KRS_qb = gdf_KRS.merge(qb_KRS, on='ARS', how='left')

gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"] = gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"].fillna(0)
gdf_KRS_qb["Max_Anteil_I71_GC_Fallzahlen"] = gdf_KRS_qb["Max_Anteil_I71_GC_Fallzahlen"].fillna(0)

gdf_KRS_qb
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# 2. Farbdefinition
# -1 → Weiß (kein GC-Zentrum)
# 0 → Grau (GC-Zentrum ohne I71-Fälle)
# >0 → Farbstufung
vmin, vmax = gdf_KRS_qb[gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"] > 0]["Min_Anteil_I71_GC_Fallzahlen"].min(), \
             gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"].max()

# Norm und Farbkarte
norm = mcolors.Normalize(vmin=vmin, vmax=vmax)
cmap = plt.cm.OrRd

# 3. Plot vorbereiten
fig, ax = plt.subplots(1, 1, figsize=(12, 10))

# b) Mit GC-Zentrum, aber 0% I71 (grau)
gdf_KRS_qb[gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"] == 0].plot(
    ax=ax, color="lightgrey", edgecolor="black", linewidth=0.5
)

# a) Ohne GC-Zentrum (grey)
gdf_KRS_qb[gdf_KRS_qb["Gefaesschirurgische_Fachabteilungen"] == 0].plot(
    ax=ax, color="grey", edgecolor="black", linewidth=0.5
)

# c) Mit GC-Zentrum und I71-Fällen (stetige Farbskala)
gdf_KRS_qb[gdf_KRS_qb["Min_Anteil_I71_GC_Fallzahlen"] > 0].plot(
    ax=ax,
    column="Min_Anteil_I71_GC_Fallzahlen",
    cmap='plasma',
    norm=norm,
    edgecolor="black",
    linewidth=0.5,
    legend=True,
    legend_kwds={
        'label': "Min. Anteil von I71-Hauptdiagnosen an GC-Fallzahlen [in %]",
        'shrink': 0.6,
        'orientation': "vertical"
    }
)

# # Punkte plotten
# qb.plot(
#     ax=ax,
#     color='black',
#     markersize=15,
#     alpha=0.8,
#     edgecolor ="white",
#     linewidth=0.5
# )
# 4. Titel & Achsen
ax.set_title("Verteilung des minimalen Anteils I71 an GC-Fallzahlen je ARS", fontsize=15)
ax.axis("off")

# 5. Eigene Legende für Grau/Weiß
custom_patches = [
    mpatches.Patch(color='lightgrey', label='GC-Zentrum ohne I71-Fälle'),
    mpatches.Patch(color='grey', label='Kein GC-Zentrum')
]
ax.legend(handles=custom_patches, loc='upper right', title='Sonderfälle', fontsize='small', title_fontsize='medium', bbox_to_anchor=(1.3, 0.17))

# 6. Speichern
plt.savefig("Gemeinsamer_Bundesausschluss/processed/Abbildungen/Min_Anteil_I71_KRS.png", dpi=300, bbox_inches='tight')
plt.show()
```
